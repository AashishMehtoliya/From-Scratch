```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston_data = load_boston()
```

```python
print(boston_data.keys())
```

```python
print(boston_data.DESCR)
```

```python
boston = pd.DataFrame(boston_data.data , columns = boston_data.feature_names)
boston.head()
```

```python
boston['MEDV'] = boston_data.target
boston.head()
```

```python
boston.describe()
```

```python
boston.isnull().sum()
```

```python
sns.set(rc = {'figure.figsize': (10 , 6)})
sns.distplot(boston['MEDV'] , bins =30)
```

```python
corr = boston.corr()
sns.heatmap(data = corr , annot = True)
```

```python
plt.figure(figsize = (20 ,5))
features = ["LSTAT" , "RM"]

for i , col in enumerate(features):
    plt.subplot(1 , len(features) , i+1)
    plt.scatter(boston[col] , boston["MEDV"] , s = 10)
    plt.xlabel(col)
    plt.ylabel("MEDV")
    plt.title(col)
```

```python
X = boston[["LSTAT" , "RM"]]
y = boston[["MEDV"]]

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score

regressor = LinearRegression()
regressor.fit(X_train , y_train)
```

```python
#accuracy of model
y_predicted_test = regressor.predict(X_test)
rmse_train = (np.sqrt(mean_squared_error(y_test , y_predicted_test)))
rsquared_train = r2_score(y_test , y_predicted_test)

print("MSE : " , rmse_train)
print("R2 : " , rsquared_train)
```

```python
#accuracy of model
y_predicted_train = regressor.predict(X_train)
rmse_train = (np.sqrt(mean_squared_error(y_train , y_predicted_train)))
rsquared_train = r2_score(y_train , y_predicted_train)

print("MSE : " , rmse_train)
print("R2 : " , rsquared_train)
```
